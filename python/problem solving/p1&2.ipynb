```python
#problem 1
# reverse word 
def reverse_words(s):
   l = s.split()
   l.reverse()
   s = " ".join(l)
   return s
```

```python
def are_you_playing_banjo(name):
    if name[0] == 'r' or name[0] == 'R':
        return f"{name} plays banjo"
    else:
        return f"{name} does not play banjo"


are_you_playing_banjo("samuel") 
```
