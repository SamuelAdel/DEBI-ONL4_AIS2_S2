```python
def user_info(name : str, age : int):
    """
    this function takes a user's name and age as input and returns a formatted string.

    :param_1 : take the name of user from input user interface
    type param_1 : str
    :param_2 : take the age of user from input user interface
    type param_2 : int

    :return: formatted string with user's name, age, and city
    :rtype: str,int 
    """
    return f"my name is {name} , my age is {age} years old."
```
