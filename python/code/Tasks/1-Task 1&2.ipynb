Task-1 
Suppose you have this email address “Amit_ml@gmail.edu”                                                                              - Input Validation: Check if the input string contains exactly one "@" symbol and 
at least one "." after the "@" symbol. If it's not a valid email, return "Invalid 
email".                                                                                                                               - Extract Username: Extract and return the part of the email before the "@" 
symbol.                                                                                                                              - Extract Domain: Extract and return the domain (the part between "@" and the 
last ".").                                                                                                                           - Check for Domain Ending: Check if the email ends with ".com". If it does, 
return "Commercial Domain". If it ends with ".edu", return "Educational Domain". 
Otherwise, return "Other Domain".

```python
email = input("Enter your email: ")

if email.count("@") != 1:
    print("Invalid email")
else:
    at_pos = email.find("@")


    if "." in email[:at_pos]:
        print("Invalid email")
    else:
        username = email.split("@")[0]
        domain = email.split("@")[1].split(".")[0]

        if email.endswith(".com"):
            domain_type = "Commercial Domain"
        elif email.endswith(".edu"):
            domain_type = "Educational Domain"
        else:
            domain_type = "Other Domain"

        print("Username:", username)
        print("Domain:", domain)
        print("Domain Type:", domain_type)
```

Task-2: 
Encoded Message: 
###!!@mocleW EPGTQ!!!6789 
Steps to Decode: 
1. Extract the core part of the message: "mocleW EPGTQ". 
2. Reverse the first word: "mocleW" becomes "Welcome". 
3. Replace shifted vowels in the second word: 
     o "EPGTQ": No vowels to change. 
4. Final decoded message: "Welcome PGTQ".

```python
encoded_message = "###!!@mocleW EPGTQ!!!6789"

core = ""
for ch in encoded_message:
    if ch.isalpha() or ch == " ":
        core += ch

# core: "mocleW EPGTQ"

words = core.split()

first_word = words[0][::-1]

second_word = words[1]

decoded_message = first_word + " " + second_word

print(decoded_message)
```

Task-3: 
Encoded Message: 
&&&**$gnirtS PLIO!!@1234 
Steps to Decode: 
1. Extract the core part of the message: "gnirtS PLIO". 
2. Reverse the first word: "gnirtS" becomes "String". 
3. Replace shifted vowels in the second word: 
    o "PLIO": Replace I->E and O->U to get "PLEU". 
4. Final decoded message: "String PLEU". 


```python
encoded_message = "&&&**$gnirtS PLIO!!@1234"

core = ""
for ch in encoded_message:
    if ch.isalpha() or ch == " ":
        core += ch

# core: "gnirtS PLIO"

words = core.split()

first_word = words[0][::-1]

second_word = ""
for ch in words[1]:
    if ch == "I":
        second_word += "E"
    elif ch == "O":
        second_word += "U"
    else:
        second_word += ch

decoded_message = first_word + " " + second_word

print(decoded_message)
```

Task-4: 
Encoded Message: 
##$$$@!yalpstcejorp EPUVT****9887 
Steps to Decode: 
1. Extract the core part of the message: "yalpstcejorp EPUVT". 
2. Reverse the first word: "yalpstcejorp" becomes "projectplay". 
3. Replace shifted vowels in the second word: 
o "EPUVT": Replace E->A, U->O to get "APOVT". 
4. Final decoded message: "projectplay APTOV".

```python
encoded_message = "##$$$@!yalpstcejorp EPUVT****9887"

core = ""
for ch in encoded_message:
    if ch.isalpha() or ch == " ":
        core += ch

# core: "yalpstcejorp EPUVT"

words = core.split()

first_word = words[0][::-1]

second_word = ""
for ch in words[1]:
    if ch == "E":
        second_word += "A"
    elif ch == "U":
        second_word += "O"
    else:
        second_word += ch

decoded_message = first_word + " " + second_word

print(decoded_message)
```
