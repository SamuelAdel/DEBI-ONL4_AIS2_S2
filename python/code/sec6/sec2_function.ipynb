```python
def factorial_1(num):
    if num == 0:
        return 1
    else:
        return num * factorial_1(num - 1)
```

```python
factorial_1(5)
```

```python
def is_prime(num : int):
    """
    
    """
    if num <= 1:
        return False
    for i in range(2, int(num)):
        if num % i == 0:
            return False
    return True
```

```python
is_prime(3)
```

```python
def reverse_txt(t):
    """"
    Reverses the words in a given string.
    """
    split_txt = t.split(" ")
    new_list = split_txt[::-1]
    Result = " ".join(new_list)
    return Result
```

```python
reverse_txt("hello world")
```

```python
def password_generator(length: int):
    """
    Generates a random password of specified length.
    """
    import random 
    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890!@#$%^&*"
    password = ""
    for _ in range(length):
        password += random.choice(characters)
    return f"Generated password: {password}"
```
