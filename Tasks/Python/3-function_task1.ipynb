```python
import os
import random

dir = "D:/git/depi/DEBI-ONL4_AIS2_S2/python/code/file_handling/fuction_task"
def thanos_project(dir, num_files):
    """
    This function creates a directory and fills it with a specified number of files.
    Then it deletes half of the files randomly.
    """
    # Create folder if not exists
    if not os.path.exists(dir):
        os.makedirs(dir)
        print("Folder created")
    else:
        print("Folder already exists")

    # Create files
    for i in range(num_files):
        file_path = os.path.join(dir, f"file_{i}.txt")
        with open(file_path, "w") as f:
            f.write("Thanos was here")

    # Check number of files
    files = os.listdir(dir)
    print("Files before deletion:", len(files))

    # Delete random half
    half = len(files) // 2
    files_to_delete = random.sample(files, half)

    for file in files_to_delete:
        os.remove(os.path.join(dir, file))

    # Check result
    final_files = os.listdir(dir)
    print("Files after deletion:", len(final_files))
```

```python
thanos_project(dir, 10)
```
