```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
x = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
y_pred_lin = lin_reg.predict(x)
print(y_pred_lin)
print(lin_reg.coef_, lin_reg.intercept_)
```

```python
# Ridge Regression
line_ridge = Ridge(alpha = 10.0)
line_ridge.fit(x, y)
y_pred_ridge = line_ridge.predict(x)
print(y_pred_ridge)
print(line_ridge.coef_, line_ridge.intercept_)
```

```python
line_lasso = Lasso(alpha = 10.0)
line_lasso.fit(x, y) 
y_pred_lasso = line_lasso.predict(x)
print(y_pred_lasso)
print(line_lasso.coef_, line_lasso.intercept_)
```

```python
print("linear Regression")
print("MSE:", mean_squared_error(y, y_pred_lin))
print("R2 Score:", r2_score(y, y_pred_lin))


print("\nRidge Regression")
print("MSE:", mean_squared_error(y, y_pred_ridge))
print("R2 Score:", r2_score(y, y_pred_ridge))


print("\nLasso Regression")
print("MSE:", mean_squared_error(y, y_pred_lasso))
print("R2 Score:", r2_score(y, y_pred_lasso))
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred_lin, color='red', label='Linear Regression')
plt.title('Linear Regression')
plt.legend()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred_ridge, color='red', label='Ridge Regression')
plt.title('Ridge Regression')
plt.legend()
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred_lasso, color='red', label='Lasso Regression')
plt.title('Lasso Regression')
plt.legend()
```

```python
plt.scatter(x, y, color = 'blue', label = 'Data Points')
plt.plot(x, y_pred_lin, color = 'red', label = 'Linear Regression')
plt.plot(x, y_pred_ridge, color = 'green', label = 'Ridge Regression')
plt.plot(x, y_pred_lasso, color = 'orange', label = 'Lasso Regression')
plt.title('Comparison of Regression Models')
plt.title('Comparison of Regression Models')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
