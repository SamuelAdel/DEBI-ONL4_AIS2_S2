```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This is the sample 2 dimensional data we are going to reduce
 
x = np.matrix([2.5,0.5,2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1  ])
 
y = np.matrix([2.4,0.7,2.9,2.2,3.0,2.7,1.6,1.1,1.6,0.9])
original_data = np.r_[x,y]
```

```python
plt.scatter([x],[y])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Original Data')
plt.show()
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
x = x - mean_x
y = y - mean_y
```

```python
plt.scatter([x],[y])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Original Data')
plt.show()
```

```python
data = np.r_[x,y]
co_variance = np.matmul(data, data.T) / (10-1)
print(co_variance)
#eigen_values,eigen_vectors = np.linalg.eig(co_variance)
```

```python
evals, evecs = np.linalg.eig(co_variance)
print("Eigen Values:\n", evals)
print("Eigen Vectors:\n", evecs)
```

```python
evecs = np.flip(evecs, axis=1)
evals = np.flip(evals)
print("Sorted Eigen Vectors:\n", evecs)
print("Sorted Eigen Values:\n", evals)
```

```python
print(np.dot(evecs[:,0].T,evecs[:,1]))
```

```python
plt.scatter([x],[y])
plt.plot([-evecs[0,0]*2,0,evecs[1,0]*2], [-evecs[1,0]*2,0,evecs[1,0]*2], color='red')
plt.plot([-evecs[0,1]*2,0,evecs[1,0]*2], [-evecs[1,1]*2,0,evecs[1,1]*2], color='yellow')
plt.axis('equal')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Eigen Vectors')
```
