```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 2.8, 3.6, 4.5])
```

```python
w = 0
b = 0
alpha = 0.01
num_iteretaion = 20
```

```python
sse_value = []
for i in range(num_iteretaion):
    y_hat = w * x + b
    #error = y_hat - y

    D_w = (2 / len(x)) * np.sum((y_hat - y) * x)
    D_b = (2 / len(x)) * np.sum(y_hat - y)

    w = w - alpha * D_w
    b = b - alpha * D_b

    sse = np.sum((y_hat - y) ** 2)
    sse_value.append(sse)

    if (i + 1) % 20 == 0:
        print(f"Iteration {i + 1}: w = {w:.4f}, b = {b:.4f}, SSE = {sse:.4f}")
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(num_iteretaion), sse_value, label='sse')
plt.xlabel('Iteration')
plt.ylabel('SSE')
plt.title('SSE over Iterations')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue', label='Data Points')
# x_line = np.linspace(min(x), max(x), 100)
# y_line = w * x_line + b
plt.plot(x, w * x + b, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Regression Line')
plt.legend()
```

```python
class LinearRegressionGD:
    def __init__(self, alpha=0.01, num_iterations=1000):
        self.alpha = alpha
        self.num_iterations = num_iterations
        self.w = 0
        self.b = 0

    def fit(self, X, y):
        n_samples, n_features = X.shape

        for _ in range(self.num_iterations):
            y_hat = np.dot(X, self.w) + self.b
            dw = (1 / n_samples) * np.dot(X.T, (y_hat - y))
            db = (1 / n_samples) * np.sum(y_hat - y)

            self.w -= self.alpha * dw
            self.b -= self.alpha * db

    def predict(self, X):
        return np.dot(X, self.w) + self.b
    
    def visualize(self, X, y):
        plt.scatter(X, y, color='blue', label='Data Points')
        x_line = np.linspace(min(X), max(X), 100)
        y_line = self.w * x_line + self.b
        plt.plot(x_line, y_line, color='red', label='Regression Line')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Data Points and Regression Line')
        plt.legend()
        plt.show()
```

```python
l = LinearRegressionGD(alpha=0.01, num_iterations=1000)
l.fit(x.reshape(-1, 1), y)
l.visualize(x, y)
```
